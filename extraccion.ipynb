```python
# Abrir el archivo de texto en modo lectura
with open('1930--uruguay/cup.txt', 'r') as file:
    # Variable para indicar si hemos encontrado algún caracter de inicio
    encontrado = False
    
    # Variable para almacenar el último caracter de inicio encontrado
    ultimo_caracter_encontrado = None
    
    # Leer cada línea del archivo
    for line in file:
        # Verificar si la línea comienza con alguno de los caracteres de inicio
        for caracterinicio in caracteres_inicio:
            if line.startswith(caracterinicio):
                # Cambiar el estado a encontrado
                encontrado = True
                ultimo_caracter_encontrado = caracterinicio
                # No procesar esta línea, solo continuar al siguiente ciclo
                break  # Salir del bucle interno si se encuentra un caracter de inicio
        
        # Si ya encontramos algún caracter de inicio, procesar y almacenar los datos
        if encontrado:
            # Verificar si la línea contiene el carácter de cierre o si llegamos al final del archivo
            if caracter_cierre in line or not line.strip():
                # Cambiar el estado a no encontrado
                encontrado = False
                ultimo_caracter_encontrado = None
                # Saltar al siguiente ciclo para continuar con la próxima línea del archivo
                continue
            
            # Dividir la línea en partes utilizando algún criterio, por ejemplo, separador ':'
            partes = line.strip().split(':')
            # Guardar los datos en el diccionario
            if len(partes) == 2:  # Verificar que la línea tenga un formato válido (clave: valor)
                clave = partes[0].strip()
                valor = partes[1].strip()
                # Almacenar los datos en un sub-diccionario correspondiente al caracter de inicio
                if ultimo_caracter_encontrado not in db1:
                    db1[ultimo_caracter_encontrado] = {}
                db1[ultimo_caracter_encontrado][clave] = valor
```

# interesante libreria (import re) tenerla en cuenta 
proporciona funciones y métodos para trabajar con expresiones regulares (regex). las cuales son patrones utilizados para buscar y manipular texto de manera flexible y poderosa. El módulo 're' en python permite utilizar expresiones regulares para realizar operaciones, buscar patrones, reemplazar texto, dividir cadenas en partes basadas en un patron, etc.


```python
import re #proporciona funciones y metodos para trabajar con expresiones regulares (regex).
# esta funcion extrae el total de goles hechos en cada partido del mundial
def contar_goles(resultado):
    # expresiones regulares para encontrar el patron del marcador (números separados por un guion)
    marcador_match = re.search(r'\d+-\d+', resultado)
    if marcador_match:
        #si se encuentra el marcado se extrae
        marcador=marcador_match.group()
        goles_local, goles_visitante = map(int,marcador.replace('(', '').replace(')', '')[0:3].split('-'))
        return goles_local + goles_visitante
    else: 
        # si no se encuentran datos validos, devuelve cero (0)
        return 0
```
